```python
#function to get the group stage data from the last 10yrs
def get_gs_matches(year): 
    web = f'https://en.wikipedia.org/wiki/{year}%E2%80%93{str(int(year[-2:]) + 1)}_UEFA_Champions_League_group_stage'
    response=requests.get(web)
    content = response.text
    soup = BeautifulSoup(content, 'lxml')
    gs_matches = soup.find_all('div', class_= 'footballbox')

    home = []
    score = []
    away = []
    
    for gs_match in gs_matches:
        #look for the elements with the class 'fhome'
        home.append(gs_match.find('th', class_='fhome').get_text())
        #look for the elements with the class 'fscore'
        score.append(gs_match.find('th', class_='fscore').get_text())
        #look for the elements with the class 'faway'
        away.append(gs_match.find('th', class_='faway').get_text())
    
    gs_dict = {'home': home, 'score': score, 'away': away}
    df_gs = pd.DataFrame(gs_dict)
    #add a column of the respective year
    df_gs['year'] = f'{year}/{str(int(year[-2:]) + 1)}'
    return df_gs
```

```python
df_gs_list = []

# Loop through a range of values
for i in strings:
  # Create a dataframe with a single column and a single row
  df = pd.DataFrame(get_gs_matches(i))
  # Append the dataframe to the list
  df_gs_list.append(df)
    
df_group_stage_history = pd.DataFrame()

for df in df_gs_list:
    df_group_stage_history = pd.concat([df_group_stage_history, df])

#clean the data for the home teams
df_group_stage_history['home'] = df_group_stage_history['home'].str.strip()
#clean the data for the away teams
df_group_stage_history['away'] = df_group_stage_history['away'].str.strip()
#replace Internazionale with Inter Milan
df_group_stage_history['home'] = df_group_stage_history['home'].replace({'Internazionale':'Inter Milan'})
df_group_stage_history['away'] = df_group_stage_history['away'].replace({'Internazionale':'Inter Milan'})
#split the goals scored by the home and away teams into two different columns
df_group_stage_history[['Team1 goals', 'Team2 goals']] = df_group_stage_history['score'].str.extract(r'(\d+)\s*–\s*(\d+)', expand=True)
df_group_stage_history

#change the datatypes to numeric
df_group_stage_history['Team1 goals'] = pd.to_numeric(df_group_stage_history['Team1 goals'])
df_group_stage_history['Team2 goals'] = pd.to_numeric(df_group_stage_history['Team2 goals'])

#change the column names
df_group_stage_history.columns = ['Team 1', 'score', 'Team 2', 'year', 'Team1 goals', 'Team2 goals']
df_group_stage_history
#covert the df_group_stage_history to a csv file
df_group_stage_history.to_csv('group_stage_history_2010_to_2022.csv')
```
